```python
soup = bs4.BeautifulSoup(req1.text)
soup
```

```python
#fetching reviews and comments
comment = soup.find_all('div', {'class':'t-ZTKy'})
for come in comment:
    print(come.get_text() + "\n\n")
```

```python
#fetching reviews
#average overall rating
rating = soup.find('div', {'class':"_2d4LTz"}).get_text();
print(rating)
```

```python
ind_rating = soup.findAll('div', {'class':'_3LWZlK _1BLPMq'})
for rate in ind_rating:
    print(rate.get_text() + "\n\n")
```
