```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('forestfires.csv')
```

```python
data
```

```python
# visualize the distribution of age using a histogram
plt.figure(figsize=(8,6))
sns.histplot(data['area'], bins=30, kde=True)
plt.title('Distribution of Burned Area')
plt.show()
```

```python
#visualize the realtionship between temprature and wind using scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data['temp'])
sns.scatterplot(data['wind'])
plt.title('Temperature vs Wind')
plt.show()
```

```python
#create a bar plot to show the counts of differnet months in the dataset
plt.figure(figsize=(8, 6))
sns.barplot(x='month',y='RH',data=data)
```

```python
#create a box plot

sns.boxplot(x='month',y='FFMC',data=data) # used to show distribution of numeric data values
```

```python
sns.displot(data['DC'],bins=20,kde=False, color='green') #displot is used for visualization on single(univariate) variable
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=data,x='day',y='DC',color='red')
```
